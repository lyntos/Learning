# Conditional Statement:
A conditional statement in Python is a way to make decisions in your code. It lets your program run different blocks of code depending on whether a condition is True or False.

Basic Idea

Python checks a condition, and:

If it’s True → it runs one block of code

If it’s False → it can run another block (optional)

# Main Types of Conditional Statements

```python
# 1. Indentation matters (very important)

# Python uses indentation (spaces) to define blocks of code.

x = 10

if x > 5:
    print("Greater than 5")  # must be indented
```

Comparison Operators Used

    == (equal to)

    != (not equal)

    > (greater than)

    < (less than)

    >= (greater than or equal)

    <= (less than or equal)

```python

# 1. if statement

# Executes code only if a condition is true:

age = 18

if age >= 18:
    print("You are an adult")
```

```python

# 2. if-else statement

# Provides an alternative if the condition is false:

age = 16

if age >= 18:
    print("Adult")
else:
    print("Minor")
```

```python
# 3. if-elif-else statement

# Checks multiple conditions:

score = 85

if score >= 90:
    print("Grade A")
elif score >= 80:
    print("Grade B")
else:
    print("Grade C")
```

```python
# 🔹 Logical Operators

# You can combine conditions:

age = 20
has_id = True

if age >= 18 and has_id:
    print("Allowed entry")
```

```python
# Nested conditionals

# You can put an if inside another if.

age = 20
has_id = True

if age >= 18:
    if has_id:
        print("You can enter")
```

```python
# Short-hand (ternary) conditional

# A one-line version of if-else:

age = 18
status = "Adult" if age >= 18 else "Minor"
print(status)
```

```python
# Truthy and Falsy values

# In Python, some values are automatically considered False:

# 0
# None
# "" (empty string)
# [], {} (empty collections)

# Everything else is usually True.

name = ""

if name:
    print("Name exists")
else:
    print("Name is empty")
```

```python
# Combining multiple conditions

# Using and, or, and not:

x = 5

if x > 0 and x < 10:
    print("Between 1 and 9")
```

```python
# Membership conditions

# Check if something is in a list/string:

fruits = ["apple", "banana"]

if "apple" in fruits:
    print("Apple is there")
```

```python
# Pass statement

# Sometimes you need an empty block:

if True:
    pass  # placeholder, does nothing
```

```python
# Example 1: if with pass
x = 10

if x > 5:
    pass  # do nothing for now
else:
    print("x is 5 or less")

print("Program continues...")
# The if block is executed but does nothing because of pass.
```

```python
# Example 2: Using pass as a placeholder for future code
age = 15

if age < 18:
    pass  # will add code later for minors
else:
    print("Adult")

print("Check completed")
# You can write the if structure first and fill in logic later.
```

```python
# Match-case (Python 3.10+)
# What is match-case?

# match-case is a new control flow statement in Python 3.10+ that works like a switch-case in other languages.
# It is used to compare a value against multiple cases and execute the matching block.

# Makes code cleaner and easier to read than many if-elif-else statements.

# A cleaner alternative to multiple elif:

day = "Tuesday"

match day:
    case "Monday":
        print("Start of week")
    case "Friday":
        print("Almost weekend")
    case _:
        print("Regular day")
```

# While Loop

A while loop in Python is used to repeat a block of code as long as a condition
 is True.

🔹 Simple definition

“A while loop keeps running again and again until the condition becomes false.”

🔹 Basic syntax

while condition:

 code to repeat

```python
# Example 1: Basic counting
x = 1

while x <= 5:
    print(x)
    x += 1
```

```python
# Example 2: Countdown
x = 5

while x > 0:
    print(x)
    x -= 1

# 👉 Output:

# 5 4 3 2 1
```

```python
# Example 3: Infinite loop (important concept)
while True:
    print("This runs forever")

# ⚠️ Runs forever unless stopped manually.
```

```python
# Example 4: Using break

# Stops the loop early:

x = 1

while x <= 10:
    if x == 5:
        break
    print(x)
    x += 1

# 👉 Output:

# 1 2 3 4
```

```python
# Example 5: Using continue

# Skips one step:

x = 0

while x < 5:
    x += 1
    if x == 3:
        continue
    print(x)

# 👉 Output:

# 1 2 4 5
```

```python
# Example 6: User input loop
password = ""

while password != "1234":#“Keep running the loop as long as the password is NOT '1234'”
    password = input("Enter password: ")

print("Access granted")
```

# What is forward filling (ffill)?

Forward fill means replacing missing values with the last valid
(non-missing) value that appeared above it.

It’s often used in time series or ordered data.


Think of it like this:

If a value is missing, just copy the previous value down.

```python
# Simple DataFrame
import pandas as pd
import numpy as np

data = {
    'Year': [2015, 2015, 2015, 2016, 2016, 2016],
    'Return': [0.5, np.nan, 0.7, np.nan, 1.2, np.nan]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
```

```python
# Step 2: Apply forward fill
df.ffill(inplace=True)
print("\nAfter forward filling:")
print(df)
```

```python
# Example 2: Forward fill with strings
data = {'City': ['NY', None, None, 'LA', None]}
df = pd.DataFrame(data)
df.ffill(inplace=True)
print(df)
```

# Backward Fill (bfill) in Python

Backward fill is a method to fill missing values (NaN) in a DataFrame or Series by using the next valid value below it.

It’s called “backward” because it looks down the column to fill missing entries.
Often used in time series or ordered data when you want missing values to take the next known value.

```python
# Example 1: Numbers
import pandas as pd
import numpy as np

df = pd.DataFrame({'Return': [0.5, np.nan, 0.7, np.nan, 1.2, np.nan]})
print("Original DataFrame:")
print(df)

df.bfill(inplace=True)
print("\nAfter backward fill:")
print(df)
```

# What is Interpolation?

Interpolation is a method to fill missing values (NaN) in a dataset by
estimating them based on surrounding values.

Instead of just copying the previous (ffill) or next (bfill) value,
interpolation calculates a value “in between” the known values.

Commonly used for numeric or time series data.

Types of Interpolation in Pandas

Linear (default) → estimates missing values assuming a straight line between
 known points

Time-based → estimates based on time index

Others → polynomial, spline, etc.

```python
# Example: Linear Interpolation
import pandas as pd
import numpy as np

# Sample data
df = pd.DataFrame({
    'Month': [1, 2, 3, 4, 5],
    'Sales': [100, np.nan, 150, np.nan, 200]
})

print("Original DataFrame:")
print(df)

```

```python
# Interpolate missing values
df['Sales'] = df['Sales'].interpolate()

print("\nAfter interpolation:")
print(df)
```

What is Polynomial Interpolation?

Polynomial interpolation estimates missing values using a polynomial curve that fits the known data points.
Unlike linear interpolation (a straight line), it can follow curves or trends in the data.

Useful when your data is non-linear or has a trend that isn’t straight.
In Pandas, you can use interpolate(method='polynomial', order=n) where n is the degree of the polynomial.

```python
# Example: Polynomial Interpolation (2nd degree)
import pandas as pd
import numpy as np

# Sample data
df = pd.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [1, np.nan, 9, np.nan, 25]
})

print("Original DataFrame:")
print(df)

```

```python
# Polynomial interpolation of order 2
df['y'] = df['y'].interpolate(method='polynomial', order=2)

print("\nAfter polynomial interpolation:")
print(df)
```

```python
# How it works
# Missing values are estimated using a quadratic curve (2nd-degree polynomial) that passes through the known points: (1,1), (3,9), (5,25)
# Row 1 (NaN) → 4
# Row 3 (NaN) → 16

# Notice how it follows the quadratic trend instead of a straight line.
```

```python
# What is dropping missing values with thresholds?

# Sometimes, a row or column has too many missing values.
# Instead of dropping every row with any missing value, you can set a threshold:

# Keep rows/columns only if they have at least N non-missing values
# Drop rows/columns that don’t meet the threshold

# Pandas uses the dropna() method with the thresh parameter:

df.dropna(thresh=N, axis=0)  # rows
df.dropna(thresh=N, axis=1)  # columns
thresh=N → minimum number of non-NaN values required to keep the row/column
axis=0 → apply on rows, axis=1 → apply on columns
```

```python
# Example 1: Dropping rows with threshold
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'A': [1, np.nan, 3, np.nan],
    'B': [4, np.nan, np.nan, 8],
    'C': [np.nan, np.nan, 9, 10]
})

print("Original DataFrame:")
print(df)

# Drop rows that have less than 2 non-NaN values
df_cleaned = df.dropna(thresh=2, axis=0)

print("\nAfter dropping rows with less than 2 non-NaN values:")
print(df_cleaned)

# Row 1 is dropped because it only had 0 non-NaN values
```

```python
# Example 2: Dropping columns with threshold
# Drop columns that have less than 3 non-NaN values
df_cleaned_cols = df.dropna(thresh=3, axis=1)

print("\nAfter dropping columns with less than 3 non-NaN values:")
print(df_cleaned_cols)
# Column A, B, and C all have less than 3 non-NaN values, so all are dropped
```

# Key Points
thresh=N → minimum number of non-missing values required
axis=0 → check rows, axis=1 → check columns
Useful when your dataset has some incomplete rows/columns but you want to keep mostly complete data

# What is Concatenation?

Concatenation is the process of combining two or more DataFrames or Series into
 a single one.


You can concatenate vertically (stack rows) or horizontally (add columns)

Use the pd.concat() function in Pandas

```python
# Syntax
# import pandas as pd
# pd.concat([df1, df2], axis=0)  # vertical (default)
# pd.concat([df1, df2], axis=1)  # horizontal

# axis=0 → stack rows (one below the other)
# axis=1 → add columns (side by side)
# ignore_index=True → re-index the resulting DataFrame
```

```python
# Example 1: Vertical Concatenation (stack rows)
import pandas as pd

df1 = pd.DataFrame({
    'Name': ['Alice', 'Bob'],
    'Age': [25, 30]
})

df2 = pd.DataFrame({
    'Name': ['Charlie', 'David'],
    'Age': [35, 40]
})

# Vertical concatenation
df_vertical = pd.concat([df1, df2], axis=0, ignore_index=True)
print(df_vertical)
# Rows from df2 are stacked below df1
```

```python
# Example 2: Horizontal Concatenation (add columns)
df3 = pd.DataFrame({
    'City': ['NY', 'LA']
})

# Horizontal concatenation
df_horizontal = pd.concat([df1, df3], axis=1)
print(df_horizontal)

# Columns from df3 are added side by side with df1
```

```python
# 🔹 Key Points
# Use pd.concat([df1, df2]) to combine DataFrames
# axis=0 → vertical (rows), axis=1 → horizontal (columns)
# ignore_index=True → reset the index in the combined DataFrame
# Works with DataFrames or Series
```

# What is Join in Pandas?

A join combines two DataFrames based on a common column (key), similar to joins
 in SQL.


You use it when you have related data in different tables and want to merge
them together.

The common column(s) is called a key.

Types of Joins


inner	Keeps only rows with keys present in both DataFrames

left	Keeps all rows from the left DataFrame, adds matching from right

right	Keeps all rows from the right DataFrame, adds matching from left

outer	Keeps all rows from both DataFrames, fills missing with NaN

🔹 1. merge() in Pandas
Used to combine two DataFrames based on a key column(s).
Works like SQL joins: inner, left, right, outer.
Syntax:
pd.merge(df1, df2, on='key_column', how='inner')  
You can specify which column(s) to join on using on=
Can join on multiple columns

Example:

pd.merge(df1, df2, on='ID', how='left')

🔹 2. join() in Pandas

Used to combine DataFrames based on their index by default.
Simpler syntax if you want to join on row labels (index).
Can also join on columns if you use set_index() first.
df1.join(df2, how='left')  # joins using index of df1 and df2
join() is basically a shortcut for merge on index
Supports left, right, inner, outer

```python
# Example: Join DataFrames
import pandas as pd

# Left DataFrame
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie']
})

# Right DataFrame
df2 = pd.DataFrame({
    'ID': [2, 3, 4],
    'City': ['LA', 'NY', 'SF']
})

# Inner join on ID
df_inner = pd.merge(df1, df2, on='ID', how='inner')
print("Inner Join:")
print(df_inner)
# Only IDs 2 and 3 appear because they exist in both DataFrames
```

```python
# Left Join Example
df_left = pd.merge(df1, df2, on='ID', how='left')
print("\nLeft Join:")
print(df_left)
# All rows from df1 are kept. Missing matches from df2 are NaN
```

```python
import pandas as pd

# Left DataFrame
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie']
})

# Right DataFrame
df2 = pd.DataFrame({
    'ID': [2, 3, 4],
    'City': ['LA', 'NY', 'SF']
})

df_right = pd.merge(df1, df2, on='ID', how='right')
print("Right Join:")
print(df_right)
# ID 4 is kept from df2 even though it doesn’t exist in df1 → Name is NaN
```

```python
# 2️ Outer Join

# Keeps all rows from both DataFrames and fills missing values with NaN.

df_outer = pd.merge(df1, df2, on='ID', how='outer')
print("Outer Join:")
print(df_outer)
```

```python
# Example: merge vs join
import pandas as pd

# DataFrames
df1 = pd.DataFrame({'ID':[1,2,3], 'Name':['Alice','Bob','Charlie']})
df2 = pd.DataFrame({'ID':[2,3,4], 'City':['LA','NY','SF']})

# Using merge
df_merge = pd.merge(df1, df2, on='ID', how='inner')
print("Merge result:\n", df_merge)

# Using join (set index first)
df1_indexed = df1.set_index('ID')
df2_indexed = df2.set_index('ID')
df_join = df1_indexed.join(df2_indexed, how='inner')
print("\nJoin result:\n", df_join)
```
