```python
# 4. Stacked Bar Chart
# 📌 Use:

# Show total + breakdown

import matplotlib.pyplot as plt

categories = ['A', 'B', 'C']
val1 = [10, 20, 15]
val2 = [5, 10, 8]

plt.bar(categories, val1, label='Part 1')
plt.bar(categories, val2, bottom=val1, label='Part 2')

plt.legend()
plt.title("Stacked Bar Chart")
plt.show()
```

```python
# 5. Percentage (Normalized) Bar Chart
# 📌 Use:

# Show proportions instead of raw values

import numpy as np
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C']
val1 = np.array([10, 20, 15])
val2 = np.array([5, 10, 8])

total = val1 + val2

plt.bar(categories, val1/total, label='Part 1')
plt.bar(categories, val2/total, bottom=val1/total, label='Part 2')

plt.title("Percentage Bar Chart")
plt.legend()
plt.show()
```

```python
explanation:
Your Data
categories = ['A', 'B', 'C']

val1 = [10, 20, 15]
val2 = [5, 10, 8]
Step 1: Total for each category
A → 10 + 5 = 15
B → 20 + 10 = 30
C → 15 + 8 = 23
📈 Step 2: Convert into proportions
val1/total    val2/total

A → 10/15 = 0.67   5/15 = 0.33
B → 20/30 = 0.67   10/30 = 0.33
C → 15/23 = 0.65   8/23 = 0.35

👉 Each bar is now a percentage of 1 (or 100%)

📊 Step 3: What the plot shows
For each category (A, B, C):
🟦 Bottom part → val1/total
🟩 Top part → val2/total
🧠 Visual interpretation
Category A
|███████████████|  → Part 1 (67%)
|███████        |  → Part 2 (33%)
Total = 100%
Category B
|███████████████|  → Part 1 (67%)
|███████        |  → Part 2 (33%)
Total = 100%
Category C
|█████████████  |  → Part 1 (65%)
|████████       |  → Part 2 (35%)
Total = 100%
🎯 Key Idea of the Code
This line:
plt.bar(categories, val1/total)

👉 Creates the bottom half (Part 1 proportion)

This line:
plt.bar(categories, val2/total, bottom=val1/total)

👉 Places Part 2 on top of Part 1

So each bar becomes:

Part 1 + Part 2 = 1 (100%)
📊 Final Output Meaning

You are NOT comparing raw values anymore.

Instead, you're showing:

“What percentage of each category is Part 1 vs Part 2?”

🧠 Why this is useful

This chart helps you:

Compare composition
See ratios instead of raw numbers
Understand how each category is divided
⚠️ Important Insight

Even though:

B has larger values (20, 10)
A has smaller values (10, 5)

👉 In the chart, both may look similar because we normalized them to 100%

🔥 One-line Summary

👉 This chart shows how each category is split into proportions of 100%, not actual values.
```

```python
# 3. Histogram
# 📌 When to use:
# Show distribution of data
# Example:
import matplotlib.pyplot as plt

data = [1, 2, 2, 3, 3, 3, 4]

plt.hist(data, bins=4)
plt.title("Histogram")

plt.show()
```

```python
You are creating a Histogram.

👉 A histogram shows:

“How many times values appear in different ranges (bins)”

📦 Your data
[1, 2, 2, 3, 3, 3, 4]
📊 Step 1: What is a histogram doing?

Instead of plotting each value separately, it:

👉 Groups values into intervals (bins)
👉 Then counts how many values fall in each bin

📦 Step 2: bins=4 means

Your data range:

Min = 1
Max = 4

So it divides into 4 intervals (bins):

Bin 1 → 1 to 1.75
Bin 2 → 1.75 to 2.5
Bin 3 → 2.5 to 3.25
Bin 4 → 3.25 to 4
📊 Step 3: Count values in each bin
Your data:
1, 2, 2, 3, 3, 3, 4
🟦 Bin-wise grouping:
Bin 1 (1 to 1.75)
1 → ✔
👉 Count = 1
Bin 2 (1.75 to 2.5)
2, 2 → ✔✔
👉 Count = 2
Bin 3 (2.5 to 3.25)
3, 3, 3 → ✔✔✔
👉 Count = 3
Bin 4 (3.25 to 4)
4 → ✔
👉 Count = 1
📊 Final histogram output (what you see)
Bin 1 → █
Bin 2 → ██
Bin 3 → ███
Bin 4 → █
🧠 What the chart is telling you

👉 Most values are in range around 3
👉 Fewer values at extremes (1 and 4)

🎯 Key idea

A histogram does NOT plot exact values.

Instead it shows:

“How data is distributed across ranges”
```

# What is a Box Plot?

A box plot (box-and-whisker plot) shows the distribution of data and helps

you quickly see:

Median (middle value)

Spread of data

Outliers (unusual values)

Key Parts of a Box Plot

Min ---- Q1 ---- Median ---- Q3 ---- Max

Minimum → smallest value

Q1 (25%) → lower quartile

Median (50%) → middle value

Q3 (75%) → upper quartile

Maximum → largest value

👉 The “box” is between Q1 and Q3

👉 The “line inside” is the median

```python
# Basic Box Plot Example
import matplotlib.pyplot as plt

data = [10, 20, 30, 40, 50]

plt.boxplot(data)
plt.title("Simple Box Plot")

plt.show()

# output
# Key values:
# Minimum (Min) = 10
# Q1 (25%) = 20
# Median (Q2) = 30
# Q3 (75%) = 40
# Maximum (Max) = 50
```

```python
Explanation of Each Part

1. The Box
Starts at Q1 (20)
Ends at Q3 (40)
Represents the middle 50% of data

2. The Line Inside the Box
At 30
This is the median (middle value)

3. The Whiskers (Lines Outside Box)
Left whisker → from 20 to 10
Right whisker → from 40 to 50

👉 They show the spread of the remaining data

4. Outliers
None in this dataset
👉 Because all values are evenly spaced

Final Interpretation
Data is perfectly evenly distributed
Median is exactly in the center
No skew (balanced data)
No outliers

Your box plot tells you:

👉 “The data ranges from 10 to 50, with the middle values between 20 and 40, and the center at 30.”
```

# What Does “Skewed Data” Mean?

Skewed data means the values are not evenly distributed.

Right skew (positive skew) → tail goes to the right

Left skew (negative skew) → tail goes to the left

```python
# 👉 1. Right-Skewed Data (Positive Skew)
# Example:
import matplotlib.pyplot as plt

data = [10, 12, 15, 18, 20, 22, 100]  # 100 pulls it right

plt.boxplot(data)
plt.title("Right Skewed Data")

plt.show()
# What Happens in the Output?
# Median shifts toward the lower values
# Right whisker becomes longer
# 100 may appear as an outlier

# Conceptual Shape:
# [ 12 ── 15 ── 20 ]─────────────── 100
#    box        ↑ long right tail

# 👉 Interpretation:
# Most values are small, but a few large values stretch the data to the right.
```

```python
# 2. Left-Skewed Data (Negative Skew)
# Example:
import matplotlib.pyplot as plt

data = [5, 40, 45, 50, 55, 60, 65]  # 5 pulls it left

plt.boxplot(data)
plt.title("Left Skewed Data")

plt.show()

# What Happens in the Output?
# Median shifts toward higher values
# Left whisker becomes longer
# 5 may appear as an outlier
# Conceptual Shape:
# 5 ───────────────[ 45 ── 50 ── 55 ]
#       long left tail       box

# 👉 Interpretation:
# Most values are high, but a few small values stretch the data to the left.
```

```python
# Example with Realistic Data
import matplotlib.pyplot as plt

marks = [45, 50, 55, 60, 65, 70, 75, 80, 90]

plt.boxplot(marks)
plt.title("Student Marks Distribution")

plt.show()

# Your box plot tells you:

# 👉 “The data ranges from 45to 90, with the middle values between 55 and 75, and the center at 65.”
```

```python
# What is IQR?

# IQR stands for Interquartile Range.

# It is a measure of how spread out the middle 50% of your data is.

# 🧠 Simple Definition

# 👉 IQR = Q3 − Q1

# Where:

# Q1 (First Quartile) = 25% of data below this value
# Q3 (Third Quartile) = 75% of data below this value

# So IQR tells you:

# “How wide the middle half of the data is”

# 📊 Example

# Data:

# [10, 20, 30, 40, 50]
# Step 1: Find quartiles
# Q1 = 20
# Q3 = 40
# Step 2: Compute IQR
# IQR = Q3 - Q1
# IQR = 40 - 20 = 20
# 🎯 Meaning of IQR
# Small IQR → data is tightly packed
# Large IQR → data is more spread out
# 🚨 Why IQR is Important
# 1. Detecting Outliers

# We use this rule:

# Lower limit = Q1 - 1.5 × IQR
# Upper limit = Q3 + 1.5 × IQR

# Any value outside this range is considered an outlier.

# 2. More Robust than Range

# Unlike range (max − min), IQR:

# Ignores extreme values
# Focuses on “typical data”
# 📦 Visual Idea (Box Plot Connection)

# In a box plot (like in Matplotlib):

# |----[ Q1 ==== Q2 ==== Q3 ]----|
#          ↑ IQR (box size)

# 👉 The box itself = IQR

# 🧠 Quick Summary
# IQR = spread of middle 50% of data
# Formula: Q3 − Q1
# Used to detect outliers
# Represents the box in a box plot
```

```python
# 1. Manual Method (Step-by-step)
import numpy as np

data = [10, 20, 30, 40, 50]

# Step 1: sort data
data = sorted(data)

# Step 2: find Q1 and Q3
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)

# Step 3: compute IQR
IQR = Q3 - Q1

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
```

```python
# 2. Using NumPy (Clean & Common)
import numpy as np

data = [10, 20, 30, 40, 50]

Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

print("IQR:", IQR)
```

```python
# 3. Using Pandas (Best for real datasets)
import pandas as pd

data = pd.Series([10, 20, 30, 40, 50])

Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

print("IQR:", IQR)
```

```python
# Bonus: Detect Outliers using IQR

# This is the most important real-world use:

data = [10, 20, 30, 40, 50, 100]

Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [x for x in data if x < lower_bound or x > upper_bound]
print(lower_bound)
print(upper_bound)

print("Outliers:", outliers)

# Final Interpretation
# Normal data range = -15 to 85
# Any value:
# < -15 ❌ are outliers

# 85 ❌ are outliers

# In this dataset:
# 👉 The outliers is 100

# Simple Intuition

# Think of it like a safe zone box:

#         Lower Bound        Upper Bound
#               |----------------|
#               safe normal data

# Anything outside = “unusual value”
```

# What is an Area Chart?

An area chart is a type of graph where the space under a line is filled with color.

It is used to show:

Trends over time

Total accumulation

Comparison of quantities visually


In Matplotlib, it is usually created using plt.fill_between().

🧠 Simple Idea

Think of it like a line chart + filled color underneath:

Line chart:      /\__

Area chart:     ████▇▇▇▇▆▆▆▆ (filled below line)

```python
# Basic Area Chart Example
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [3, 5, 2, 6, 4]

plt.plot(x, y)                 # line
plt.fill_between(x, y)         # area fill

plt.title("Simple Area Chart")
plt.show()
```

```python
# Colored Area Chart
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [3, 5, 2, 6, 4]

plt.plot(x, y, color='blue')
plt.fill_between(x, y, color='skyblue', alpha=0.5)

plt.title("Colored Area Chart")
plt.show()
```

```python
# Area Chart from Zero Baseline (Important)

# Most real area charts fill from 0 to line, not just between points.

plt.fill_between(x, y,0, color='green', alpha=0.4)
plt.plot(x, y, color='green')
plt.title("Area Chart from Zero Baseline")
plt.show()
```

```python
# Stacked Area Chart (Advanced)

# Used to compare multiple series:

y1 = [2, 3, 4, 3, 5]
y2 = [1, 2, 2, 3, 2]

plt.stackplot(x, y1, y2, labels=['A', 'B'], alpha=0.7)
plt.legend()
plt.title("Stacked Area Chart")
plt.show()
```
