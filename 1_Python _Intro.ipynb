# Bitwise Operators

In Python, bitwise operators are used to perform operations on the binary

(bit-level) representation of integers

They work directly on bits (0 and 1).

These operators are commonly used in low-level programming, cryptography, and performance optimization.

image.png


```python
a = 5   # binary: 0101
b = 3   # binary: 0011

print(a & b)  # 1  (0001)
print(a | b)  # 7  (0111)
print(a ^ b)  # 6  (0110)
```

image.png

Step 1: Convert numbers to binary
a = 5  → 0101  
b = 3  → 0011
🔹 1. a & b (AND operator)

👉 Rule: 1 & 1 = 1, otherwise 0

  0101
& 0011
------
  0001  → 1

✔ Output: 1

🔹 2. a | b (OR operator)

👉 Rule: if any bit is 1 → result is 1

  0101
| 0011
------
  0111  → 7

✔ Output: 7

🔹 3. a ^ b (XOR operator)

👉 Rule: 1 if bits are different, 0 if same

  0101
^ 0011
------
  0110  → 6

✔ Output: 6

💡 Easy Rules to Remember:
Operator	Name	Rule
&	AND	Both must be 1
|	OR	At least one is 1
^	XOR	Different bits → 1
🔁 Simple way to think:

AND (&) → common 1s

OR (|) → combine all 1s

XOR (^) → only different 1s

```python
a = 5
print(~a)   # -6
```

Step 1: Convert a to binary
a = 5  # binary: 0000 0101
Step 2: Apply ~ (bitwise NOT)

Flip all bits: 0 → 1, 1 → 0

0000 0101  → 1111 1010

In Python, numbers are signed integers, so it uses two’s complement.

Formula: ~a = -(a + 1)

~5 = -(5 + 1) = -6
✅ Output:
-6
💡 Easy way to remember:

~x = -(x + 1)

```python
# Shift Operators

a = 5  # binary: 0101

print(a << 1)  # 10 (1010)
print(a >> 1)  # 2  (0010)
```

Step 1: Binary of a
a = 5  # binary: 0101
🔹 1. Left Shift (<<)

Syntax: a << n → shift bits n positions to the left

Effect: multiplies the number by 2 for each shift

0101 << 1 → 1010

Binary 1010 = 10 in decimal

✔ Output: 10

🔹 2. Right Shift (>>)

Syntax: a >> n → shift bits n positions to the right

Effect: divides the number by 2 (floor division) for each shift

0101 >> 1 → 0010

Binary 0010 = 2 in decimal

✔ Output: 2

💡 Easy way to remember:

<< left shift → multiply by 2

>> right shift → divide by 2

# Slicing

Slicing in Python is a technique used to extract a part (subset) of a sequence.

In Python, slicing is a way to access a portion (subsequence) of a sequence
like a list, tuple, or string instead of just one element. Think of it as
cutting out a segment of your data

It is applied to sequence data types such as:

    •	String
    •	List
    •	Tuple
    •	Range
Syntax

sequence[start : stop : step]

•	start → starting index

•	stop → ending index (not included)

•	step → interval between elements


```python
# Example with a List
devices = ["Laptop", "Tablet", "Smartphone", "Smartwatch", "Monitor"]

print(devices[1:4])  # ['Tablet', 'Smartphone', 'Smartwatch']
print(devices[:3])   # ['Laptop', 'Tablet', 'Smartphone'] → start default 0
print(devices[2:])   # ['Smartphone', 'Smartwatch', 'Monitor'] → till end
print(devices[::2])  # ['Laptop', 'Smartphone', 'Monitor'] → every 2nd element
```

```python
# Example with a Tuple
components = ("CPU", "GPU", "RAM", "SSD", "HDD")

print(components[1:4])  # ('GPU', 'RAM', 'SSD')
```

```python
# Example with a String
device = "MacBookPro"

print(device[0:3])  # 'Mac'
print(device[3:])   # 'BookPro'
print(device[::-1]) # 'orPkbooBcaM' → reversed string
```

Slicing lets you extract part of a sequence.


start:stop → from start index to stop-1 index


step → how many elements to skip


Works for lists, tuples, and strings

```python
t=(10,20,40,50,60,45,76)
print(t[0:5:2])
```

1️⃣ bool(0) → False

Any numeric zero (0, 0.0) evaluates to False.

2️⃣ bool(25) → True

25 is a non-zero number, so it is truthy.

Python treats empty containers (like "", [], {}) as False.

4️⃣ bool("Hi") → True

"Hi" is a non-empty string, which is truthy.

Any non-empty container (string, list, tuple, dict, etc.) evaluates to True.

# Order of Operations in Python

When an expression contains multiple mathematical operations, Python follows a
 specific order to decide which operation to perform first.

Python follows the PEMDAS rule.

Python solves expressions from left to right when operations have the same priority.


image.png

```python
# Example 1

print(2 + 3 * 4)

# Step-by-step:
# 1.	Multiplication first → 3 * 4 = 12
# 2.	Addition → 2 + 12 = 14
```

```python
# Example 2 (Using Parentheses)

print((2 + 3) * 4)

# Step-by-step:
# 1.	Parentheses first → 2 + 3 = 5
# 2.	Multiply → 5 * 4 = 20
```

```python
# Example 3 (Exponent)

print(2 + 3 ** 2)

# Step-by-step:
# 1.	Exponent → 3 ** 2 = 9
# 2.	Addition → 2 + 9 = 11
```

# Sorting and Filtering Data in Python

When working with data sets, we often need to:

Sorting means arranging data in order (ascending or descending).

Python provides the sorted() function to sort lists.


•	Sort data → arrange it in order

Filtering means selecting only the elements that meet a condition.

Example: selecting even numbers only.


•	Filter data → select only the values we want

These techniques help us analyze and understand data better


```python
# Sorting in Ascending Order
numbers = [3, 1, 4, 1, 5, 9, 2, 6, 5]

sorted_numbers = sorted(numbers)

print("Sorted Numbers:", sorted_numbers)
```

```python
sorted_numbers_desc = sorted(numbers, reverse=True)

print("Sorted Numbers (Descending):", sorted_numbers_desc)
```

```python
# Filtering Data
# Filtering means selecting only the elements that meet a condition.
# Example: selecting even numbers only.
# ________________________________________
# Method 1: Using List Comprehension (Most Common)

numbers = [3, 1, 4, 1, 5, 9, 2, 6, 5]

even_numbers = [num for num in numbers if num % 2 == 0]

print("Even Numbers:", even_numbers)


# •	num % 2 == 0 → checks if the number is even
# •	Only numbers that satisfy the condition are included

```

What is a Lambda Function?

A lambda function is a small anonymous function written in one line.

It is used for performing quick function

Example:

lambda x: x % 2 == 0

Meaning:

Take x → return True if x is even


```python
# Using filter() Function
# Python also has a filter function.

numbers = [3, 1, 4, 1, 5, 9, 2, 6, 5]

even_numbers_filter = list(filter(lambda x: x % 2 == 0, numbers))

print("Even Numbers (Using filter):", even_numbers_filter)
```
